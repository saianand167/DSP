```python
marks = [
    [78, 85, 90],
    [88, 76, 92],
    [69, 80, 75],
    [90, 95, 85]
]

# 1. Total marks of each student
total_marks = [sum(student) for student in marks]
print("Total marks of each student:", total_marks)

# 2. Average marks per subject
num_students = len(marks)
num_subjects = len(marks[0])

avg_per_subject = [
    sum(marks[i][j] for i in range(num_students)) / num_students
    for j in range(num_subjects)
]
print("Average marks per subject:", avg_per_subject)

# 3. Find the topper (highest total)
topper_index = total_marks.index(max(total_marks))
print("Topper (student index):", topper_index)
print("Topper total marks:", total_marks[topper_index])
```
