```python
import re
text = "apple and banana are awesome animals"
matches = re.findall(r"\b[aA]\w*", text)
print(matches)

# \b is a word boundary 
# [aA] matches either 'a' or 'A'
# \w* matches zero or more word characters (letters, digits, or underscores) following the initial 'a' or 'A'.
```

```python
# i will find word starts with r
text="raj is as good as rani"
matches=re.findall(r"\b[rR]\w*",text)
print(matches)
```

```python
# 7.Write a RegEx to match one or more lowercase letters. 
import re 
prtext = "This is a test string with LowerCase letterS."
matches = re.findall(r"\b[a-z]+\w+", prtext)
print(matches)
```

```python
# 8.Write a RegEx to match a word ending with ‘ing’. 
text="This is a test string for running and jumping."
matches=re.findall(r"\b\w+ing",text)
print(matches)
matches2=re.findall(r"\b\w+ing",text)
print(matches2)
```

```python
# 9.Write a RegEx to match alphanumeric strings only.
import re
text = "this 215yygy iuiu	1234 abc_123"
matches=re.findall(r"\b[a-zA-Z0-9_]+\b",text)
print(matches)
```

```python
# 10. Write a RegEx to validate a 10-digit mobile number starting with 
# 6–9.
import re
text = "sai bhudu 9876544456 686785535 989886786875 98909077987"
matches=re.findall(r"\b[6-9]\d{9}\b",text)
print(matches)
```

```python
import re

email = "sai%@gmail.com"

pattern = r"^[A-Za-z0-9._\%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"

if re.fullmatch(pattern, email):
    print("Valid email address")
else:
    print("Invalid email address")
```
