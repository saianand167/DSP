```python
import re
text="hi i am raj"
print(re.search(r"raj",text)) #it  will find raj in the text and return a match object
print(re.search(r"raj",text).group()) #it will return the matched string
```

```python
text ="good good study, day day up"
list=re.findall(r"good",text)
print(list)
print(re.findall(r"good",text, re.IGNORECASE)) #it will ignore the case and find all the good in the text
```

```python
import re

text = "Raj and Raj are friends"
matches = re.finditer(r"Raj", text)


for match in matches:
    print("Found:", match.group(), "at index", match.start())
```
