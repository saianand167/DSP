```python
import pandas as pd

ages = [5,6, 12, 18, 25, 32, 40, 55]
bins = [0, 10, 20, 30, 40, 50, 60]
labels = ["0-10", "11-20", "21-30", "31-40", "41-50", "51-60"]

binned = pd.cut(ages, bins=bins,	labels=labels, right=False)
print(binned)
```

```python
import pandas as pd

scores = [10, 50, 78, 92, 30, 66, 55]

bins = [0, 25, 50, 75, 100]
labels = ["Fail", "Pass", "Good", "Excellent"]

grades = pd.cut(scores, bins=bins, labels=labels)

print(grades)
```
