```python
# write mode delete old files and create new file
file=open("a.txt","w")
file.write("hello world\nwelcome to dsp ay25-26")
# print(file.read()) # will give error because file is in write mode
file.close()
```

```python
file=open("a.txt","r")
# we can read now
print(file.read())
file.close()
```

```python
# append mode add new data to existing file
file=open("a.txt","a")
file.write("\ni can read	and write")
file.close()
```

```python
# r+ mode we can read and write both
file=open("a.txt","r+")
# print(file.read())
file.write("this is r+4 mode\n")
file.close() 
# data will be added at the end of the file default data will not be clear
```

```python
# w+ 
file=open("c.txt","w+")
file.write("this is w+ mode\n")
file.write("old data will be deleted\n")
print(file.read()) # will give empty because pointer is at the end after write
file.close()
```

```python
# a+
file=open("a.txt","a+")
file.write("this is a+ mode\n")
print(file.read()) # will give empty because pointer is at the end after write
file.close()
```

```python
import sys
import time
import numpy as np

# Create a list of 1,000,000 integers
list_1000000 = list(range(1_000_000))

# Create a NumPy array of 1,000,000 integers
array_1000000 = np.arange(1_000_000)

# Print memory usage
print(f"Size of list: {sys.getsizeof(list_1000000)} bytes")
print(f"Size of array: {array_1000000.nbytes} bytes")

# Time summing the list
start_time = time.time()
sum_list = sum(list_1000000)
end_time = time.time()
list_sum_time = (end_time - start_time) * 1000

# Time summing the NumPy array
start_time = time.time()
sum_array = np.sum(array_1000000)
end_time = time.time()
array_sum_time = (end_time - start_time) * 1000
print(sum_array)
print(sum_list)
# Print results
print(f"Time to sum list: {list_sum_time:.2f} milliseconds")
print(f"Time to sum array: {array_sum_time:.2f} milliseconds")
```
